```python
#We will use spacy's EDS-NLP for this. It's the only French clinical NER model I could find off-the-shelf - nearly all of them work only on English.
import spacy
import os
from tqdm import tqdm
```

```python
# Declare the pipeline
nlp = spacy.blank("eds")

# General-purpose components
nlp.add_pipe("eds.normalizer")
nlp.add_pipe("eds.sentences")

#Miscellaneous pipelines
nlp.add_pipe("eds.dates")
nlp.add_pipe("eds.measurements")

# Entity extraction pipelines - we will ignore scores/codes because these are unlikely to crop up in research papers (rather than clinical notes), and may do more harm than good
# We are interested in terminology, that being drugs, names, diseases, etc.
nlp.add_pipe("eds.drugs")
nlp.add_pipe("eds.cim10", config=dict(term_matcher="simstring"))
nlp.add_pipe("eds.covid")

#Qualifier pipelines - need to adapt for medical research papers, so negation and hypotheses only (we won't see patient history etc)
nlp.add_pipe("eds.negation")
nlp.add_pipe("eds.hypothesis")
```

```python
nlp.add_pipe("eds.CKD")
nlp.add_pipe("eds.COPD")
nlp.add_pipe("eds.cerebrovascular_accident")
nlp.add_pipe("eds.congestive_heart_failure")
nlp.add_pipe("eds.connective_tissue_disease")
nlp.add_pipe("eds.dementia")
nlp.add_pipe("eds.diabetes")
nlp.add_pipe("eds.hemiplegia")
nlp.add_pipe("eds.leukemia")
nlp.add_pipe("eds.liver_disease")
nlp.add_pipe("eds.lymphoma")
nlp.add_pipe("eds.myocardial_infarction")
nlp.add_pipe("eds.peptic_ulcer_disease")
nlp.add_pipe("eds.peripheral_vascular_disease")
nlp.add_pipe("eds.solid_tumor")
nlp.add_pipe("eds.alcohol")
nlp.add_pipe("eds.tobacco")
```

```python
dir_path = os.getcwd()
f = open(os.path.join(dir_path, "../wmt22gold.txt"), "r", encoding = "utf8")
fr_sent = [line.strip() for line in f.readlines()]
f.close()
```

```python
for sentence in fr_sent:
    print(sentence)
    doc = nlp(sentence)
    print(doc.ents)
```
