Why are some terms predicted more often than others? We'll try to answer this question here.

```python
#First, tokenise the FR side of the entire training set to obtain word frequencies. We'll obtain model-dependent subword tokens later.
from datasets import load_dataset, Dataset

#Converts data in src [TAB] tgt [NEWLINE] format to a format suitable for model training
def convertToDictFormat(data):
    source = []
    target = []
    for example in data:
        example = example.strip()
        sentences = example.split("\t")
        source.append(sentences[0])
        target.append(sentences[1])
    ready = Dataset.from_dict({"en":source, "fr":target})
    return ready
```
